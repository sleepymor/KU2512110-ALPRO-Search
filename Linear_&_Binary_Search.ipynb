```python
def linear_search(arr, target):
  """
  Melakukan pencarian linear pada sebuah list.

  Args:
    arr: List yang akan dicari.
    target: Nilai yang ingin dicari.

  Returns:
    Indeks dari target jika ditemukan, None jika tidak ditemukan.
  """
  for i in range(len(arr)):
    if arr[i] == target:
      return i
  return None

# Contoh penggunaan
my_list = [10, 20, 30, 40, 50]
target_value = 30

result = linear_search(my_list, target_value)

if result is not None:
  print(f"Nilai {target_value} ditemukan pada indeks {result}")
else:
  print(f"Nilai {target_value} tidak ditemukan dalam list")

target_value_2 = 60
result_2 = linear_search(my_list, target_value_2)

if result_2 is not None:
  print(f"Nilai {target_value_2} ditemukan pada indeks {result_2}")
else:
  print(f"Nilai {target_value_2} tidak ditemukan dalam list")
```

```python
def binary_search(arr, target):
  """
  Melakukan pencarian biner pada sebuah list yang sudah diurutkan.

  Args:
    arr: List yang sudah diurutkan yang akan dicari.
    target: Nilai yang ingin dicari.

  Returns:
    Indeks dari target jika ditemukan, None jika tidak ditemukan.
  """
  low = 0
  high = len(arr) - 1
  while low <= high:
    mid = (low + high) // 2
    if arr[mid] == target:
      return mid
    elif arr[mid] < target:
      low = mid + 1
    else:
      high = mid - 1
  return None

# Contoh penggunaan
my_sorted_list = [10, 20, 30, 40, 50]
target_value = 40

result = binary_search(my_sorted_list, target_value)

if result is not None:
  print(f"Nilai {target_value} ditemukan pada indeks {result}")
else:
  print(f"Nilai {target_value} tidak ditemukan dalam list")

target_value_2 = 60
result_2 = binary_search(my_sorted_list, target_value_2)

if result_2 is not None:
  print(f"Nilai {target_value_2} ditemukan pada indeks {result_2}")
else:
  print(f"Nilai {target_value_2} tidak ditemukan dalam list")
```
